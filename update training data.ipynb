```python
import pandas as pd
```

```python
existing_data = pd.read_csv('data_files/multisub_5_partitions_unique_old.csv')
new_data = pd.read_csv('data_files/new_data.csv')
```

```python
def reduce_size(sequence, max_size):
    if len(sequence) > max_size:
        sequence = sequence[:max_size // 2] + sequence[-max_size // 2:]
    return sequence
```

```python
new_data.index = new_data.index + 28303
new_data['Kingdom'] = new_data['Kingdom'].fillna('Metazoa')
new_data['Sequence'] = reduce_size(new_data['Sequence'],1022)
```

```python
combined_data = existing_data.append(new_data)
```

```python
combined_data.to_csv('data_files/multisub_5_partitions_unique.csv', index=False)
```

```python
new_fasta_entries = []

for index, row in new_data.iterrows():
    acc = row['ACC']
    sequence = row['Sequence']
    new_fasta_entries.append(f'>{acc}\n{sequence}')
```

```python
print(new_fasta_entries[:10])
```

```python
with open('data_files/deeploc_swissprot_clipped1k.fasta', 'a') as fasta_file:
    for entry in new_fasta_entries:
        fasta_file.write('\n' + entry)

with open('data_files/deeploc_swissprot_clipped4k.fasta', 'a') as fasta_file:
    for entry in new_fasta_entries:
        fasta_file.write('\n' + entry)
```
